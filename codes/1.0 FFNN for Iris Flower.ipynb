```python
from sklearn.datasets import load_iris

iris=load_iris()
data=iris.data
target=iris.target

from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,
                                                              test_size=0.1)
```

```python
from keras.layers import Dense #fully connected layers
from keras.models import Sequential

model=Sequential()
#an empty Neural Network

model.add(Dense(8,input_dim=4,activation='relu'))
#1st Hidden Layer
model.add(Dense(8,input_dim=8,activation='relu'))
#2nd Hideen Layer
model.add(Dense(3,input_dim=8,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
print(train_target[:5])
```

```python
from keras.utils import np_utils

new_train_target=np_utils.to_categorical(train_target)
print(new_train_target[:5])
```

```python
history=model.fit(train_data,new_train_target,epochs=1000)
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['accuracy'])
```
